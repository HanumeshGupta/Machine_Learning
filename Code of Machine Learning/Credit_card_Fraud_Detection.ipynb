This data is highly unbalanced

0--> Normal Transaction

1--> Fraudulent Transaction

```python
#Seperating the data for analysis 

ledgit = credit_data[credit_data.Class ==0]
fraud = credit_data[credit_data.Class ==1]
```

```python
print("Shape of Ledgit data: ",ledgit.shape)
print("Shape of Fraud data: ",fraud.shape)
```

Here Method we will use for making the data ready for sampling is  sample function . Here we have 492 fraud data so we gonna use that data and make same number of legit data.

```python
ledgit_sample = ledgit.sample(n= 492)
```

```python
new_dataset = pd.concat([ledgit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Now spliting the data into label data
x = new_dataset.drop(columns= 'Class' ,axis=1)
y=new_dataset['Class']
```

```python
#Traing dataset and testing dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Checking model in training data
train_pred = model.predict(x_train)
accuracy = accuracy_score(train_pred,y_train)
print("Accuracy Score of training data: ",accuracy)
```

```python
#Checking model in testing data
test_pred = model.predict(x_test)
accuracy = accuracy_score(test_pred,y_test)
print("Accuracy Score of training data: ",accuracy)
```
