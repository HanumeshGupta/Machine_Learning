```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r'D:\Visual Studio Code\ML\DataSet\breast_cancer_data.csv')
data.head()
```

```python
data.replace({'diagnosis':{'M':0 , 'B': 1}},inplace=True)
```

```python
data =data.drop(columns=['Unnamed: 32','id'],axis=1)
```

```python
data.groupby('diagnosis').mean()
```

```python
x= data.drop(columns='diagnosis',axis=1)
y = data['diagnosis']
```

```python
#splitting the data into training and testing data

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#Model Training
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Checking the model in training data

train_pred = model.predict(x_train)
accuracy = accuracy_score(y_train,train_pred)
print("Accuracy of training data: ",accuracy)
```

```python
#Checking the model in testing data

test_pred = model.predict(x_test)
accuracy = accuracy_score(y_test,test_pred)
print("Accuracy of training data: ",accuracy)
```
