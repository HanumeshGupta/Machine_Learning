```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv(r'D:\Visual Studio Code\ML\DataSet\movies.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
selected_data = ['genres','keywords','tagline','crew','director','title']
```

```python
for feature in selected_data:
    data[feature] = data[feature].fillna('')
```

```python
#Combaining all the selected feature:

feature_Data = data['genres']+data['keywords']+data['tagline']+data['crew']+data['director']+['title']
```

```python
feature_Data.head()
```

```python
#Converting the text data into feature data
vectorizer = TfidfVectorizer()
```

```python
feature_extraction = vectorizer.fit_transform(feature_Data)
```

```python
print(feature_extraction)
```

# Cosine Similarity

```python
# Getting the similarity scores using cosine Similarity

similarity = cosine_similarity(feature_extraction)
```

```python
similarity.shape
```

```python
#Getting the name of the movie from user
movie_name = input("Enter the name of the movie: ")
```

```python
list_of_movies  = data['title'].to_list()
print(list_of_movies)
```

```python
#finding rhe close match for the movie name given by the user
findind_close_match = difflib.get_close_matches(movie_name,list_of_movies)
print(findind_close_match)
```

```python
close_match = findind_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = data[data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
#Sorting the movie based on there similarity score

sorted_similar_movie = sorted(similarity_score,key = lambda x:x[1], reverse=True)
print(sorted_similar_movie)
```

```python
#Print the name of similar movie based on the index 

print('Movies suggested for you: \n')
i=1

for movie in sorted_similar_movie:
    index = movie[0]
    title_from_index = data[data.index==index]['title'].values[0]
    if (i<30):
        print(i,'.', title_from_index)
        i+=1
```
